```python
import os
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

import seaborn as sns
import matplotlib.pyplot as plt
os.chdir('C:\\Users\\prudi\\Desktop\\Data Sets\\House Pricing Dataset\\housing-prices-dataset')
hp=pd.read_csv('HousePricingDataset.csv')
#os.chdir('/home/exe002/Desktop/DataSet/housing-prices-dataset')
#hp=pd.read_csv('train.csv')


tips=sns.load_dataset('tips')
flights=sns.load_dataset('flights')
```

#### Single Categorical Variables

```python
sns.countplot(hp['Street'])

print(hp['Street'].value_counts())
```

#### Single Continuous Variables

```python
plt.figure(figsize = (10,5))
sns.distplot(hp['SalePrice']);
#skewness and kurtosis
print("Skewness: %f" % hp['SalePrice'].skew())
print("Kurtosis: %f" % hp['SalePrice'].kurt())
```

```python
sns.set_style("whitegrid")
g = sns.boxplot(x=hp['SalePrice'])
g = (g.set(xlim=(0,1000000)))
```
