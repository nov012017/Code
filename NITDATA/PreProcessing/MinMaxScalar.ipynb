```python
import pandas as pd
import numpy as np
```

```python
## Correlation and Covariance
temperature=[98,87,90,85,95,75]
numberofcustomers=[15,12,10,10,16,7]

## Creating the dataframe from the above two lists
icecream = pd.DataFrame(list(zip(temperature, numberofcustomers)), columns =['temperature', 'numberofcustomers'])
icecream
```

```python
#Number of Customers Column
from sklearn.preprocessing import MinMaxScaler #Zscore

MinMax_numberofcustomers=MinMaxScaler()

MinMax_numberofcustomers.fit(np.array(icecream['numberofcustomers']).reshape(-1, 1))

MinMax_numberofcustomers_VALUES=MinMax_numberofcustomers.transform(np.array(icecream['numberofcustomers']).reshape(-1, 1))

icecream['zscore_numberofcustomers']=MinMax_numberofcustomers_VALUES

icecream
```

```python
#Number of Customers Column
from sklearn.preprocessing import MinMaxScaler #Zscore

MinMax_numberofcustomers=MinMaxScaler()

MinMax_numberofcustomers.fit(np.array(icecream['temperature']).reshape(-1, 1))

MinMax_numberofcustomers_VALUES=MinMax_numberofcustomers.transform(np.array(icecream['temperature']).reshape(-1, 1))

icecream['zscore_temperature']=MinMax_numberofcustomers_VALUES

icecream
```
