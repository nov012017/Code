### Hypothesis Testing

```python
hp.head()
```

```python
# Apply this set of code for the ordinal variable
from sklearn.preprocessing import LabelEncoder
le_HeatingQC = LabelEncoder()
le_HeatingQC.fit(hp['HeatingQC'])
le_HeatingQC=pd.DataFrame(le_HeatingQC.transform(hp['HeatingQC']),columns=['HeatingQC_Transformed'])
hp= pd.concat([hp, le_HeatingQC],axis=1)
hp.head()
```

### Chi square test

```python
## Performing Chi square between two input columns 'HeatingQC_Transformed' AND 'Alley_Transformed'

from sklearn.feature_selection import chi2
f_score,pvalue = chi2(np.array(hp['Alley_Transformed']).reshape(-1,1), np.array(hp['HeatingQC_Transformed']).reshape(-1,1))
print('P-Value for this is: ',pvalue)
print('F score for this is: ',f_score)

#HO: Independent
#HA: DEPENDENT

# As we got very high pvalue, we will be accepting the HO.
# Which means as we performed chi-square test between two input variables. We can conclude that these two are not dependent on
# each other. Hence we can keep these two columns
```

### T-TEST

```python
from scipy.stats import ttest_ind
stat, pvalue = ttest_ind(hp['LotArea'], hp['SalePrice'])
print('P-Value for this is: ',pvalue)
print('t stat for this is: ',stat)

#HO: Both are same
#HA: Both are not same

# As we got very small pvalue, we will be rejecting the HO.
# Which means as we performed ttest between one input and output variables. We can conclude that these two are not dependent on
# each other. Hence we can remove 'LotArea' from the dataset
```
