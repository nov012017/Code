```python
import nltk
```

```python
paragraph=" The tajmahal was built by experor shah jahan"
```

```python
words=nltk.word_tokenize(paragraph)
```

```python
words
```

```python
tagged_words=nltk.pos_tag(words)
```

```python
tagged_words
```

```python
sent=[]
for i in tagged_words:
    if i[1]!='NN':
        sent.append(i[0])
```

```python
sent=' '.join(sent)
sent
```
