```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

os.chdir('C:\\Users\\Prudhvi\\Desktop\\Prudhvi\\Data Science\\Kaggle')
```

```python
df=pd.read_csv('habermans-survival-data-set.csv')
df.columns=['col1','col2','col3','col4']
df.head()
```

```python
print('Unique values in col4=',df.col4.unique())
pd.crosstab(index=df.col4, columns='count')
```

```python
sns.FacetGrid(df,hue='col4',size=3)\
    .map(sns.distplot,'col1')
plt.xlabel('Column 1')
plt.ylabel('Count')
plt.title('PDF of Col1 vs Col4')

sns.set_style('whitegrid');
sns.FacetGrid(df,hue='col4')\
    .map(sns.distplot,'col2')
plt.xlabel('Column 2')
plt.ylabel('Count')
plt.title('PDF of Col2 vs Col4')

sns.set_style('whitegrid');
sns.FacetGrid(df,hue='col4',size=3)\
    .map(sns.distplot,'col3')
plt.xlabel('Column 3')
plt.ylabel('Count')
plt.title('PDF of Col3 vs Col4')
```

```python
fig, ax =plt.subplots(1,3)
sns.boxplot('col4','col1',data=df,ax=ax[0])
sns.violinplot('col4','col1',data=df,ax=ax[2])
```

```python
titanic = sns.load_dataset("titanic")
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
```

```python
tips.head()
```

```python
sns.set(style="whitegrid", color_codes=True)
```

```python
#Categorical scatterplots
sns.stripplot('day','total_bill',data=tips)
```

```python
sns.stripplot('day','total_bill',data=tips,jitter=True)
```

```python
sns.swarmplot('day','total_bill',data=tips)
plt.title('DAY vs Total Bill')

sns.FacetGrid(tips,size=4)\
    .map(sns.swarmplot,'day','total_bill')
```

```python
sns.swarmplot('day','total_bill',hue='sex',data=tips,size=4)

sns.FacetGrid(tips,hue='sex',size=4)\
    .map(sns.swarmplot,'day','total_bill')
```

```python
sns.boxplot('total_bill','day','sex',data=tips,dodge=True)
```
