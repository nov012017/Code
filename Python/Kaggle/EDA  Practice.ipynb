```python
df.plot(kind='scatter',x='wheelbase',y='enginesize');
plt.show()
```

```python
sns.set_style('whitegrid');
sns.FacetGrid(df,hue='fueltype', size=4)\
    .map(plt.scatter,'wheelbase','enginesize')\
    .add_legend();
plt.show()
plt.close()
```

```python
sns.set_style("whitegrid")
sns.pairplot(df,hue='fueltype', size=3)
plt.show()
plt.close();
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(df,hue='fueltype', size=4)\
    .map(sns.distplot,'enginesize')\
    .add_legend()
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(df, size=8)\
    .map(sns.distplot,'enginesize')\
    .add_legend()
plt.show()
```

```python
counts, bin_edges=np.histogram(df['enginesize'],bins=15,density=False)
print(counts)
print('****************')
print(bin_edges)
print('****************')
print(sum(counts))
pdf = counts/(sum(counts))
print('******')
print(pdf);
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'r',label='CDF',linewidth=5.0)
plt.plot(bin_edges[1:],cdf,label='PDF')
plt.grid
plt.legend()
plt.ylabel('Counts')
plt.xlabel('Engine Size')
plt.show
```
