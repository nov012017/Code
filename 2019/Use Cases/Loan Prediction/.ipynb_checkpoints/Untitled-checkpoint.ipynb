```python
import pickle
from sklearn.preprocessing import LabelBinarizer
from collections import defaultdict
import pandas as pd
Labelencoderpickle_in = open("C:\\Users\\Administrator\\Documents\\GitHub\\Code\\2019\\Use Cases\\Loan Prediction\\lablebinariser_pickle.pkl","rb")
Labelencoder_dict = pickle.load(Labelencoderpickle_in)
def LabelBinarizer_apply(data,columns):
    for feat in nominal:
        encoded_df=pd.DataFrame()
        d=pd.DataFrame(Labelencoder_dict[feat].transform(data[feat]))
        test_column=d.columns.values
        list_columns=[feat+'_'+str(j) for j in test_column]
        #print(list_columns)
        d.columns=list_columns
        #print(d.head())
        encoded_df=pd.concat([encoded_df,d],axis=1).reset_index(drop=True)
        #data.drop(feat,axis=1,inplace=True)
        data=data.reset_index(drop=True)
        data=pd.concat([data,encoded_df],axis=1)
    return data 
#with open('C:\\Users\\prudi\\Desktop\\Data Sets\\Loan Prediction\\binarizer_pickle.pkl','wb') as f:
#    pickle.dump(encoding_binarizer,f)
dataset_preprocessed=LabelBinarizer_apply(dataset,nominal)
```

```python
from sklearn.preprocessing import MinMaxScaler
from collections import defaultdict
encoding_minmax=defaultdict(MinMaxScaler)
features_num=['ltv','AVERAGE.ACCT.AGE','CREDIT.HISTORY.LENGTH','age']
Minmaxscaler = open("C:\\Users\\Administrator\\Documents\\GitHub\\Code\\2019\\Use Cases\\Loan Prediction\\minmax_pickle.pkl","rb")
MinMaxScaling_dict = pickle.load(Minmaxscaler)
def scalingminmax_apply(data,columns):
    for feat in features_num:
        encoded_df=pd.DataFrame()
        #d=pd.DataFrame(encoding_minmax(feat).fit_transform(new_df[feat]))
        d=pd.DataFrame(MinMaxScaling_dict[feat].fit_transform(data[[feat]]))
        test_column=d.columns.values
        list_column=[feat+'_'+'minmax' for j in test_column]
        d.columns=list_column
        encoded_df=pd.concat([encoded_df,d],axis=1).reset_index(drop=True)
        #data.drop(feat,axis=1,inplace=True)
        data=data.reset_index(drop=True)
        data=pd.concat([data,encoded_df],axis=1)
    return data
dataset_preprocessed=scalingminmax_apply(dataset_preprocessed,features_num)
```
