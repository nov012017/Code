```python
import re

str = "The rain in Spain b"

#Find all lower case characters alphabetically between "a" and "m":
print(str)

x = re.findall(r'[a-d]', str)
print(x)

x=re.sub('[a-d]','[A-D]',str)
print(x)

x=re.search(r'[a-d]',str)
if x:
    print('Match Found')
else:
    print('Match Not Found')
```

```python
import re

str = "The rain in Spain 9 5 Hello 5667.1"

#Find all lower case characters alphabetically between "a" and "m":

x = re.findall(r'\d', str)
print(x)

x=re.sub(r'\d','[A-D]',str)
print(x)

x=re.search(r'\d',str)
if x:
    print('Match Found')
else:
    print('Match Not Found')
```

```python
import re

str = "hello world heiio"

#Search for a sequence that starts with "he", followed by two (any) characters, and an "o":

x = re.findall("he..o", str)
print(x)

x=re.sub(r'he..o','[A-D]',str)
print(x)

x=re.search(r'he..o',str)
if x:
    print('Match Found')
else:
    print('Match Not Found')
```

```python
import re

str = "hello world"

#Check if the string starts with 'hello':

x = re.findall("^hello", str)
print(x)


x=re.sub(r'^hello','[A-D]',str)
print(x)

x=re.search(r'^hello',str)
if x:
    print('Match Found')
else:
    print('Match Not Found')
```

```python
import re

str = "The ai12 rain in Spain falls mainly in the plain! ai"

#Check if the string contains "ai" followed by 0 or more "x" characters:

x = re.findall("ai*", str)

print(x)

if (x):
  print("Yes, there is at least one match!")
else:
  print("No match")

x=re.sub(r'ai*','[A-D]',str)
print(x)

x=re.sub(r'aix*','[A-D]',str)
print(x)
```

```python
import re

str = "The aix rain in Spain falls mainly in the plain! ai"

#Check if the string contains "ai" followed by 1 or more "x" characters:

x = re.findall("ai+", str)

print(x)

if (x):
  print("Yes, there is at least one match!")
else:
  print("No match")

x=re.sub(r'aix+','[A-D]',str)
print(x)

x=re.sub(r'ai+','[A-D]',str)
print(x)
```

```python
import re

str = "The rain in Spain falllls mainly in the plain! all"

#Check if the string contains "a" followed by exactly two "l" characters:

x = re.findall("al{2}", str)

print(x)

if (x):
  print("Yes, there is at least one match!")
else:
  print("No match")
```

### Email

```python
email='prudivds@gm8.-ail.com'
x=re.findall(r'^([a-z0-9-\.-]*)@([\da-z\.-]*)\.([a-z\.]{2,6})$',email)   # or
x=re.findall(r'^([a-z0-9-\.-]+)@([\da-z\.-]+)\.([a-z\.]{2,6})$',email)
print(x)
if x:
    print('Match Found')
else:
    print('Match Not Found')
```
